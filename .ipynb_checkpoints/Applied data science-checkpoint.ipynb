```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.stattools import adfuller

# Load the data from the CSV file and set the 'Date' column as the index
v = pd.read_csv('data.csv', index_col='Date')

# Sort the data by the index (Date)
data =v.sort_index()['Close']
print(data)
# Calculate the log percentage change
log_percentage_change =  (data.pct_change()*100).dropna()

# Plot the log percentage change
log_percentage_change.plot(figsize=(10, 6))
plt.title('Log Percentage Change Over Time')
plt.xlabel('Date')
plt.ylabel('Log Percentage Change')
plt.grid(True)
plt.show()

```

```python
adfuller(log_percentage_change)
```

```python
log_percentage_change

```

```python
from statsmodels.tsa.arima_process import arma_generate_sample
```

```python
ar_coefs=[1]
ma_coefs=[1,-0.7]
Y = arma_generate_sample(ar_coefs, ma_coefs,nsample=1000, scale=0.1)
plt.plot(Y)
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA (log_percentage_change, order=(1,0,0), exog=v['Volume'].sort_index().pct_change().dropna() )
results= model.fit()
results.summary()
```

```python
one_step = results.get_prediction(start=-30, dynamic=True)
mean = one_step.predicted_mean
confidence_intervals = one_step.conf_int()
lower_limits = confidence_intervals.loc[:,'lower Close']
upper_limits = confidence_intervals.loc[:,'upper Close']
```

```python
plt.plot(log_percentage_change.index, log_percentage_change, label='observed')
plt.plot(mean.index,mean, color='b')
```

```python
m = log_percentage_change.mean()
std = log_percentage_change.std()
print(m,std)
```

```python
import arch

model = arch.arch_model(
,p=1,q=1, mean='constant', vol='GARCH',dist='skewt')
mdlreslt = model.fit(update_freq=4)
print(mdlreslt.summary())
frcst = mdlreslt.forecast(horizon=5)
print(frcst.variance[-1:])
stdresid = mdlreslt.resid/mdlreslt.conditional_volatility
```

```python
plt.hist(stdresid, facecolor='orange')
```

```python
plt.plot(stdresid)
plt.plot(log_percentage_change)
plt.show()
```

```python
import os
import pandas as pd
import glob as g 

all_files = g.glob('C:/Users/zgcha/FTD/data/*/*')

# Define the directory where your folders are located
data_directory = 'C:/Users/zgcha/FTD/data'

concatenated_df = pd.DataFrame()

#glob, os.lisdir() os.path.join()


for folder in os.listdir(data_directory):
    folder_path = os.path.join(data_directory, folder)
    
    # Check if it's a directory (not a file)
    if os.path.isdir(folder_path):
        # Loop through files in the folder
        for file in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file)
            
            # Check if it's a file (not a subdirectory)
            if os.path.isfile(file_path):
                # Read the data file into a DataFrame
                df = pd.read_excel(file_path)  # You may need to adjust the read function based on your file format
                # Concatenate the DataFrame to the main DataFrame
                concatenated_df = pd.concat([concatenated_df, df], ignore_index=True)

# Now, concatenated_df contains all the data from your folders
concatenated_df.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
concatenated_df.dropna()
concatenated_df = concatenated_df.replace(' ', pd.NA).dropna() # or to_numeric(df.col, errors='coerce')
concatenated_df['Value'] = concatenated_df['Value'].astype(float)
result_df = concatenated_df.groupby('Date')['Value'].mean()
print(result_df)
concatenated_df['Date'] = pd.to_datetime(concatenated_df['Date'])

concatenated_df.set_index('Date', inplace=True)
monthly_df = concatenated_df.resample('M').last()
descriptive_stats = monthly_df['Value'].describe()
print(descriptive_stats)
```
